```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python

data = {
    'X1': [1, 2, 6, 3, 10, 5],
    'X2': [4, 2, 4, 5, 6, 11]
}

df = pd.DataFrame(data)


scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

```

Hierarchical Clustering (Hiyerarşik Kümeleme):

```python
from scipy.cluster.hierarchy import linkage, dendrogram

linkage_matrix = linkage(scaled_data, 'ward')
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering, DBSCAN


dbscan = DBSCAN(eps=1, min_samples=2)
df['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_data)


plt.scatter(df['X1'], df['X2'], c=df['DBSCAN_Cluster'], cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
from sklearn.cluster import MeanShift


mean_shift = MeanShift(bandwidth=1)
df['MeanShift_Cluster'] = mean_shift.fit_predict(scaled_data)


plt.scatter(df['X1'], df['X2'], c=df['MeanShift_Cluster'], cmap='viridis')
plt.title('Mean Shift Clustering')
plt.show()
```
